```python
import numpy as np
```

# 5 students, 4 subjects: Maths, Science, English, History

```python
marks = np.array([
    [88, 72, 65, 90],
    [55, 48, 70, 60],
    [92, 85, 88, 95],
    [70, 60, 75, 68],
    [40, 55, 50, 45]
])
```

```python
students = ["Aarav", "Priya", "Ravi", "Sneha", "Arjun"]
subjects = ["Maths", "Science", "English", "History"]
```

# 1. Average marks per student (each row)

```python
avg_per_student = marks.mean(axis=1)
```

# 2. Average marks per subject (each column)

```python
avg_per_subject = marks.mean(axis=0)
```

# 3. Who passed? (average >= 60)

```python
passed = np.array(students)[avg_per_student >= 60]
```

# 4. Normalise marks to 0–1 range

```python
normalised = (marks - marks.min()) / (marks.max() - marks.min())
```

# 5. Apply grade bonus of 3 marks to Maths only (broadcasting)

```python
bonus = np.array([3, 0, 0, 0])
adjusted = marks + bonus
```
